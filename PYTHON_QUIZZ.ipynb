QUIZZ SUR PYTHON, DU PLUS FACILE AU PLUS DIFFICILE

```python
def ask_question(question, options, correct_option):
    print(question)
    for i, option in enumerate(options, 1):
        print(f"{i}. {option}")

    answer = input("Veuillez entrer le numéro de votre réponse : ")

    try:
        answer = int(answer)
        if answer == correct_option:
            print("Bonne réponse !\n")
            return True
        else:
            print(f"Mauvaise réponse. La bonne réponse était : {options[correct_option - 1]}\n")
            return False
    except ValueError:
        print("Entrée invalide. Veuillez entrer un numéro.\n")
        return False

def main():
    questions = [
        {
            "question": "Quel est le type de données pour un nombre entier en Python ?",
            "options": ["int", "str", "float", "bool"],
            "correct_option": 1
        },
        {
            "question": "Quelle est la sortie de 'print(2 ** 3)' ?",
            "options": ["5", "6", "8", "9"],
            "correct_option": 3
        },
        {
            "question": "Comment déclarez-vous une liste en Python ?",
            "options": ["()", "[]", "{}", "<>"],
            "correct_option": 2
        },
        {
            "question": "Comment commencez-vous un commentaire en Python ?",
            "options": ["#", "//", "/*", "--"],
            "correct_option": 1
        },
        {
            "question": "Quelle méthode est utilisée pour ajouter un élément à une liste ?",
            "options": [".append()", ".add()", ".insert()", ".push()"],
            "correct_option": 1
        },
        {
            "question": "Quelle est la sortie de 'print(type('hello'))' ?",
            "options": ["<class 'str'>", "<class 'int'>", "<class 'list'>", "<class 'dict'>"],
            "correct_option": 1
        },
        {
            "question": "Comment créez-vous une fonction en Python ?",
            "options": ["function myFunction() {}", "def myFunction():", "create myFunction()", "function:myFunction()"],
            "correct_option": 2
        },
        {
            "question": "Quelle est la sortie de 'len([1, 2, 3, 4])' ?",
            "options": ["3", "4", "5", "6"],
            "correct_option": 2
        },
        {
            "question": "Comment accédez-vous au troisième élément d'une liste 'myList' ?",
            "options": ["myList[2]", "myList[3]", "myList(2)", "myList(3)"],
            "correct_option": 1
        },
        {
            "question": "Quelle est la sortie de 'print(5 / 2)' ?",
            "options": ["2", "2.5", "2.0", "2.25"],
            "correct_option": 2
        },
        {
            "question": "Comment vérifiez-vous l'égalité de deux valeurs 'a' et 'b' en Python ?",
            "options": ["a = b", "a == b", "a != b", "a equal b"],
            "correct_option": 2
        },
        {
            "question": "Quelle est la méthode correcte pour convertir une chaîne en majuscules ?",
            "options": ["str.upper()", "str.capitalize()", "str.toUpper()", "str.upcase()"],
            "correct_option": 1
        },
        {
            "question": "Quel mot-clé est utilisé pour gérer les exceptions en Python ?",
            "options": ["try", "except", "catch", "finally"],
            "correct_option": 2
        },
        {
            "question": "Quelle est la sortie de 'print(10 % 3)' ?",
            "options": ["0", "1", "2", "3"],
            "correct_option": 2
        },
        {
            "question": "Comment vérifiez-vous si une clé existe dans un dictionnaire 'myDict' ?",
            "options": ["'key' in myDict", "'key' exists in myDict", "'key' is in myDict", "has_key('key') in myDict"],
            "correct_option": 1
        },
        {
            "question": "Quel est le mot-clé pour définir une classe en Python ?",
            "options": ["class", "def", "struct", "object"],
            "correct_option": 1
        },
        {
            "question": "Quelle est la sortie de 'print(bool(0))' ?",
            "options": ["True", "False", "None", "Error"],
            "correct_option": 2
        },
        {
            "question": "Quelle méthode est utilisée pour obtenir une vue des clés dans un dictionnaire ?",
            "options": ["myDict.keys()", "myDict.get_keys()", "myDict.key_list()", "myDict.get()"],
            "correct_option": 1
        },
        {
            "question": "Quel mot-clé est utilisé pour créer une boucle en Python ?",
            "options": ["for", "loop", "iterate", "repeat"],
            "correct_option": 1
        },
        {
            "question": "Comment arrêtez-vous une boucle en Python ?",
            "options": ["stop", "exit", "break", "terminate"],
            "correct_option": 3
        }
    ]

    score = 0

    for q in questions:
        if ask_question(q['question'], q['options'], q['correct_option']):
            score += 1

    print(f"Vous avez répondu correctement à {score} question(s) sur {len(questions)}.")

if __name__ == "__main__":
    main()
```
